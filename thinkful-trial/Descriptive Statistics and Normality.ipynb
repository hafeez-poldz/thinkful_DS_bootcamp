```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# making Beta distributed variable with 100 observtions
np.random.seed(42)
beta = np.random.beta(81, 219, 100)

# sorting variable in ascending order
beta.sort()

# plotting a histogram for Beta distribution 
plt.hist(beta, bins = 20, color = 'brown')

# adding a vertical line at the mean.
plt.axvline(beta.mean(), color = 'c', linestyle = 'solid', linewidth = 2)

# adding a vertical line at one standard deviation above the mean
plt.axvline(beta.mean() + beta.std(), color = 'c', linestyle = 'dashed', linewidth = 2)

# adding a vertical line at one standard deviation below the mean
plt.axvline(beta.mean()-beta.std(), color = 'c', linestyle = 'dashed', linewidth = 2 )
plt.show()
```

```python
# Additionally:
# generating two normally-distributed variables
norm1 = np.random.normal(5, 0.5,100)
norm2 = np.random.normal(10, 1, 100)

# adding them together to create a third variable.
norm3 = norm1+norm2

# plotting the third variable using a histogram
plt.hist(norm3, bins = 20, color = 'orange')

# compute the mean and standard deviation and plot them as vertical lines on the histogram
mean = norm3.mean()
std = norm3.std()

plt.axvline(mean, color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(mean + std, color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean - std, color = 'red', linestyle = 'dashed', linewidth = 2)
plt.show()
# Evaluate the descriptive statistics against the data.
print("Results change depending on random seed")
```
